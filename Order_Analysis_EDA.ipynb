Loading necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading Dataset

```python
dataset=pd.read_csv(r'C:\DATASCIENCE\EDA\Order_Analaysis_EDA.csv')
df=pd.DataFrame(dataset)
print(df.to_string)
```

```python
df.head()
```

Removing Spaces in the names of the columns

```python
df.columns = [col.replace(' ', '_') for col in df.columns]
```

```python
df.columns
```

Dropping Unnecessary Columns

```python
drop_columns = ['Order_ID', 'Order_Date', 'Ship_Date', 'Ship_Mode', 'Customer_ID', 'Customer_Name']
df = df.drop(columns=drop_columns)
```

```python
df.info()
```

```python
df.shape
```

Changing the Datatypes

```python
new_columns=['Aging','Quantity','Discount']
df[new_columns] = df[new_columns].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

Removing the dollar$ symbol in columns- Profit,Sales,Shipping Cost

```python
df['Profit'] = df['Profit'].astype(str).str.replace(r'\$', '', regex=True).astype(float)
df['Sales'] = pd.to_numeric(df['Sales'].str.replace(r'\$', '', regex=True), errors='coerce')
df['Shipping_Cost'] = pd.to_numeric(df['Shipping_Cost'].str.replace(r'\$', '', regex=True), errors='coerce')
```

Removing the null values

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

Finding Correlation

```python
df1=df
numeric_df = df1.select_dtypes(include=['number'])

corr_matrix = numeric_df.corr()

print(corr_matrix)
```

```python
plt.figure(figsize=(10, 8))

sns.heatmap(corr_matrix, annot=True)

plt.show()
```

Finding Outlier

```python
df2=df.select_dtypes(exclude=['object'])
for column in df2:
        plt.figure(figsize=(17,1))
        sns.boxplot(data=df1, x=column)
```

No outliers Found

Plotting the relationships

```python
sns.scatterplot(x='Shipping_Cost',y='Profit',data=df)
plt.xlabel("Shipping_Cost")
plt.ylabel("Profit")
plt.show
```

```python
sns.barplot(x='Discount',y='Profit',data=df)
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show
```

```python
mean_profit_by_product = df.groupby('Product')['Profit'].mean().reset_index()
plt.figure(figsize=(10, 6)) 
sns.barplot(x='Product', y='Profit', data=mean_profit_by_product)
plt.xlabel("Product")
plt.ylabel("Profit")
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.show()
```

```python
mean_profit_by_product = df.groupby('Product_Category')['Profit'].mean().reset_index()
sns.barplot(x='Product_Category', y='Profit', data=mean_profit_by_product)
plt.xlabel("Product_Category")
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.show()
```

```python
mean_profit_by_segment = df.groupby('Segment')['Profit'].mean().reset_index()
plt.figure(figsize=(5, 5))
plt.pie(mean_profit_by_segment['Profit'], labels=mean_profit_by_segment['Segment'], autopct='%1.1f%%', startangle=90)
plt.title('Mean Profit by Segment')
plt.axis('equal')  
plt.show()
```
